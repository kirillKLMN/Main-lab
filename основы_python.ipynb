#–ó–∞–¥–∞—á–∞ 1
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ –≤—Ö–æ–¥ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö –±—É–∫–≤.
–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞—Å–Ω—ã—Ö –±—É–∫–≤ (a, e, i, o, u) –≤ —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–µ. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def count_vowels(text):
    vowels = 'aeiou'
    count = 0
    text_lower = text.lower()
    
    for char in text_lower:
        if char in vowels:
            count += 1         
    return count

import unittest

#–¢–µ—Å—Ç—ã

class TestCountVowels(unittest.TestCase):
    
    def test_empty_string(self):
        self.assertEqual(count_vowels(""), 0)
    
    def test_only_vowels(self):
        self.assertEqual(count_vowels("aeiou"), 5)
        self.assertEqual(count_vowels("aieou"), 5)
    
    def test_only_consonants(self):
        self.assertEqual(count_vowels("bcdfghjklmnpqrstvwxyz"), 0)
    
    def test_mixed_case(self):
        self.assertEqual(count_vowels("Hello World"), 3)
        self.assertEqual(count_vowels("PYTHON"), 1)
        self.assertEqual(count_vowels("PrOgrAmMing"), 3)
    
    def test_with_numbers_and_symbols(self):
        self.assertEqual(count_vowels("Hello123!@#"), 2)
        self.assertEqual(count_vowels("a1e2i3o4u5"), 5)
    
    def test_long_string(self):
        long_text = "This is a long string with many vowels for testing purposes"
        self.assertEqual(count_vowels(long_text), 16)
    
    def test_single_vowels(self):
        self.assertEqual(count_vowels("a"), 1)
        self.assertEqual(count_vowels("e"), 1)
        self.assertEqual(count_vowels("x"), 0)
```

# –ó–∞–¥–∞—á–∞ 2
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ –≤—Ö–æ–¥, –∫–æ—Ç–æ—Ä–æ–π –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å true, –µ—Å–ª–∏
–∫–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª –≤ —Å—Ç—Ä–æ–∫–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ 1 —Ä–∞–∑, –∏–Ω–∞—á–µ –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å false. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def has_unique_chars(text):
    seen_chars = set() 
    for char in text:
        if char in seen_chars:
            return False
        seen_chars.add(char)
    return True

#–¢–µ—Å—Ç—ã 
import unittest

class TestHasUniqueChars(unittest.TestCase):
    
    def test_empty_string(self):
        self.assertTrue(has_unique_chars(""))
    
    def test_single_character(self):
        self.assertTrue(has_unique_chars("a"))
        self.assertTrue(has_unique_chars("1"))
        self.assertTrue(has_unique_chars("!"))
    
    def test_all_unique(self):
        self.assertTrue(has_unique_chars("abc"))
        self.assertTrue(has_unique_chars("123"))
        self.assertTrue(has_unique_chars("a1b2"))
        self.assertTrue(has_unique_chars("python"))
    
    def test_duplicate_characters(self):
        self.assertFalse(has_unique_chars("hello"))
        self.assertFalse(has_unique_chars("112"))
        self.assertFalse(has_unique_chars("test"))
        self.assertFalse(has_unique_chars("abca"))
    
    def test_case_sensitivity(self):
        self.assertTrue(has_unique_chars("AaBb"))  
        self.assertTrue(has_unique_chars("Hello"))  
    
    def test_special_characters(self):
        self.assertTrue(has_unique_chars("!@#$"))
        self.assertFalse(has_unique_chars("!!"))
        self.assertTrue(has_unique_chars("a@b#c"))
    
    def test_spaces(self):
        self.assertTrue(has_unique_chars("a b c"))
        self.assertFalse(has_unique_chars("a  b"))  
        self.assertTrue(has_unique_chars("hello world"))  
    
    def test_long_strings(self):
        self.assertTrue(has_unique_chars("abcdefghijklmnopqrstuvwxyz"))
        self.assertFalse(has_unique_chars("abcdefghijklmnopqrstuvwxyza"))
        self.assertTrue(has_unique_chars("1234567890"))
    
    def test_unicode_characters(self):
        self.assertTrue(has_unique_chars("–ø—Ä–∏–≤–µ—Ç"))
        self.assertFalse(has_unique_chars("—Ö–æ—Ä–æ—à–æ"))  
        self.assertTrue(has_unique_chars("üéâüéäüéà"))
        self.assertFalse(has_unique_chars("üéâüéâ"))
```

# –ó–∞–¥–∞—á–∞ 3
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
–±–∏—Ç —Ä–∞–≤–Ω—ã—Ö 1 –≤ —ç—Ç–æ–º —á–∏—Å–ª–µ. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def count_bits(n):
    if n < 0:
        raise ValueError("–ß–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º") 
    count = 0
    while n > 0:
        count += n & 1  
        n >>= 1         
    return count

# –¢–µ—Ç—Å—ã 
import unittest

class TestCountBits(unittest.TestCase):
    
    def test_zero(self):
        self.assertEqual(count_bits(0), 0)
    
    def test_powers_of_two(self):
        self.assertEqual(count_bits(1), 1)      
        self.assertEqual(count_bits(2), 1)      
        self.assertEqual(count_bits(4), 1)      
        self.assertEqual(count_bits(8), 1)      
        self.assertEqual(count_bits(16), 1)     
        self.assertEqual(count_bits(32), 1)     
        self.assertEqual(count_bits(64), 1)     
        self.assertEqual(count_bits(128), 1)   
    
    def test_numbers_with_multiple_bits(self):
        self.assertEqual(count_bits(3), 2)     
        self.assertEqual(count_bits(5), 2)     
        self.assertEqual(count_bits(6), 2)     
        self.assertEqual(count_bits(7), 3)      
        self.assertEqual(count_bits(15), 4)    
        self.assertEqual(count_bits(31), 5)     
        self.assertEqual(count_bits(63), 6)    
        self.assertEqual(count_bits(127), 7)    
    
    def test_odd_numbers(self):
        self.assertEqual(count_bits(9), 2)      
        self.assertEqual(count_bits(11), 3)     
        self.assertEqual(count_bits(13), 3)   
        self.assertEqual(count_bits(17), 2)
        self.assertEqual(count_bits(21), 3)   
    
    def test_even_numbers(self):
        self.assertEqual(count_bits(10), 2)   
        self.assertEqual(count_bits(12), 2)    
        self.assertEqual(count_bits(14), 3)  
        self.assertEqual(count_bits(18), 2)     
        self.assertEqual(count_bits(20), 2)     
    
    def test_large_numbers(self):
        self.assertEqual(count_bits(255), 8)    
        self.assertEqual(count_bits(256), 1)    
        self.assertEqual(count_bits(511), 9)    
        self.assertEqual(count_bits(1023), 10)  
        self.assertEqual(count_bits(1024), 1)       
    
    def test_negative_number(self):
        with self.assertRaises(ValueError):
            count_bits(-1)
        with self.assertRaises(ValueError):
            count_bits(-10)
        with self.assertRaises(ValueError):
            count_bits(-100)
```

# –ó–∞–¥–∞—á–∞ 4
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å —Ç–æ,
—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–µ—Ä–µ–º–Ω–æ–∂–∞—Ç—å —Ü–∏—Ñ—Ä—ã —á–∏—Å–ª–∞ –∏–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–∏—è, —á—Ç–æ–±—ã
–ø–æ–ª—É—á–∏–ª–æ—Å—å —á–∏—Å–ª–æ —Å–æ—Å—Ç–æ—è—â–µ–µ –∏–∑ –æ–¥–Ω–æ–π —Ü–∏—Ñ—Ä—ã.
–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è –≤—Ö–æ–¥–Ω–æ–≥–æ —á–∏—Å–ª–∞:
¬∑ 39 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 3, —Ç–∞–∫ –∫–∞–∫ 3*9=27 => 2*7=14 => 1*4=4
¬∑ 4 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 0, —Ç–∞–∫ –∫–∞–∫ —á–∏—Å–ª–æ —É–∂–µ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –æ–¥–Ω–æ–π —Ü–∏—Ñ—Ä—ã
¬∑ 999 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 4, —Ç–∞–∫ –∫–∞–∫ 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def magic(n):
    if n < 0:
        raise ValueError("–ß–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º")
    if n < 10:
        return 0
    
    steps = 0
    current = n
    
    while current >= 10:
        digits = [int(d) for d in str(current)]
        
        product = 1
        for digit in digits:
            product *= digit
        
        current = product
        steps += 1
    
    return steps

#–¢–µ—Å—Ç—ã
import unittest

class TestMagicFunction(unittest.TestCase):
    
    def test_single_digit_numbers(self):
        self.assertEqual(magic(0), 0)
        self.assertEqual(magic(1), 0)
        self.assertEqual(magic(5), 0)
        self.assertEqual(magic(9), 0)
    
    def test_examples_from_task(self):
        self.assertEqual(magic(39), 3)  
        self.assertEqual(magic(4), 0)   
        self.assertEqual(magic(999), 4) 
    
    def test_two_digit_numbers(self):
        self.assertEqual(magic(10), 1)  
        self.assertEqual(magic(11), 1)  
        self.assertEqual(magic(25), 2)  
        self.assertEqual(magic(77), 4)  
    
    def test_three_digit_numbers(self):
        self.assertEqual(magic(100), 1)  
        self.assertEqual(magic(123), 1) 
        self.assertEqual(magic(256), 3)  
        self.assertEqual(magic(777), 4)  
```

# –ó–∞–¥–∞—á–∞ 5
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã –∏
–≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–≤—É—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("–í–µ–∫—Ç–æ—Ä—ã –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã")
    
    if len(pred) == 0:
        raise ValueError("–í–µ–∫—Ç–æ—Ä—ã –Ω–µ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –ø—É—Å—Ç—ã–º–∏")
    
    n = len(pred)
    total_error = 0.0
    for i in range(n):
        error = pred[i] - true[i]
        total_error += error ** 2
    return total_error / n

#–¢–µ—Å—Ç—ã
import unittest
import math

class TestMSEFunction(unittest.TestCase):
    
    def test_identical_vectors(self):
        self.assertEqual(mse([1, 2, 3], [1, 2, 3]), 0.0)
        self.assertEqual(mse([0, 0, 0], [0, 0, 0]), 0.0)
        self.assertEqual(mse([-1, -2, -3], [-1, -2, -3]), 0.0)
    
    def test_constant_difference(self):

        self.assertEqual(mse([2, 3, 4], [1, 2, 3]), 1.0)  
        self.assertEqual(mse([5, 6, 7], [2, 3, 4]), 9.0)  
    
    def test_varying_differences(self):

        self.assertAlmostEqual(mse([2, 2, 3], [1, 2, 2]), 2/3, places=10)
        self.assertAlmostEqual(mse([2, 4, 6], [1, 2, 3]), 14/3, places=10)
    
    def test_negative_numbers(self):
        self.assertEqual(mse([-1, -2, -3], [-2, -3, -4]), 1.0)  
        self.assertEqual(mse([1, 2, 3], [-1, -2, -3]), 56/3)   
```

# –ó–∞–¥–∞—á–∞ 6
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å
—Å—Ç—Ä–æ–∫—É –≤–∏–¥–∞ ‚Äú(n1**p1)(n2**p2)‚Ä¶(nk**pk)‚Äù –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∞—è —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ —á–∏—Å–ª–∞ –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ
–º–Ω–æ–∂–∏—Ç–µ–ª–∏ (–µ—Å–ª–∏ pi == 1, —Ç–æ –≤—ã–≤–æ–¥–∏—Ç—å —Ç–æ–ª—å–∫–æ ni).
–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —á–∏—Å–ª–∞ 86240 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å ‚Äú(2**5)(5)(7**2)(11)‚Äù. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def prime_factors(n):
    if n <= 0:
        raise ValueError("–ß–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º")
    if n == 1:
        return "(1)"
    
    factors = {}
    temp = n
    divisor = 2
    
    while divisor * divisor <= temp:
        while temp % divisor == 0:
            factors[divisor] = factors.get(divisor, 0) + 1
            temp //= divisor
        divisor += 1
    if temp > 1:
        factors[temp] = factors.get(temp, 0) + 1
    
    result_parts = []
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result_parts.append(f"({prime})")
        else:
            result_parts.append(f"({prime}**{power})")
    
    return "".join(result_parts)

#–¢–µ—Å—Ç—ã
import unittest

class TestPrimeFactors(unittest.TestCase):
    
    def test_small_numbers(self):
        self.assertEqual(prime_factors(1), "(1)")
        self.assertEqual(prime_factors(2), "(2)")
        self.assertEqual(prime_factors(3), "(3)")
        self.assertEqual(prime_factors(4), "(2**2)")
        self.assertEqual(prime_factors(5), "(5)")
        self.assertEqual(prime_factors(6), "(2)(3)")
        self.assertEqual(prime_factors(7), "(7)")
        self.assertEqual(prime_factors(8), "(2**3)")
        self.assertEqual(prime_factors(9), "(3**2)")
        self.assertEqual(prime_factors(10), "(2)(5)")
    
    def test_example_from_task(self):
        self.assertEqual(prime_factors(86240), "(2**5)(5)(7**2)(11)")
    
    def test_prime_numbers(self):
        self.assertEqual(prime_factors(11), "(11)")
        self.assertEqual(prime_factors(13), "(13)")
        self.assertEqual(prime_factors(17), "(17)")
        self.assertEqual(prime_factors(19), "(19)")
        self.assertEqual(prime_factors(23), "(23)")
        self.assertEqual(prime_factors(29), "(29)")
    
    def test_numbers_with_repeated_factors(self):
        self.assertEqual(prime_factors(12), "(2**2)(3)")
        self.assertEqual(prime_factors(18), "(2)(3**2)")
        self.assertEqual(prime_factors(24), "(2**3)(3)")
        self.assertEqual(prime_factors(36), "(2**2)(3**2)")
        self.assertEqual(prime_factors(72), "(2**3)(3**2)")
```

# –ó–∞–¥–∞—á–∞ 7
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n, –∑–∞–¥–∞—é—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—É–±–∏–∫–æ–≤. –§—É–Ω–∫—Ü–∏—è
–¥–æ–ª–∂–Ω–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –∏–∑ –¥–∞–Ω–Ω–æ–≥–æ –∫–æ–ª-–≤–∞ –∫—É–±–∏–∫–æ–≤ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–∏—Ä–∞–º–∏–¥—É, —Ç–æ –µ—Å—Ç—å
–º–æ–∂–Ω–æ –ª–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å —á–∏—Å–ª–æ n –∫–∞–∫ 1^2+2^2+3^2+‚Ä¶+k^2. –ï—Å–ª–∏ –º–æ–∂–Ω–æ, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞
–≤–µ—Ä–Ω—É—Ç—å k, –∏–Ω–∞—á–µ —Å—Ç—Ä–æ–∫—É ‚ÄúIt is impossible‚Äù. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"
    
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k ** 2
        
        if total == number:
            return k
        elif total > number:
            return "It is impossible"
    
    return "It is impossible"

#–¢–µ—Å—Ç—ã
import unittest

class TestPyramidFunction(unittest.TestCase):
    
    def test_small_valid_pyramids(self):
        self.assertEqual(pyramid(1), 1)       
        self.assertEqual(pyramid(5), 2)        
        self.assertEqual(pyramid(14), 3)       
        self.assertEqual(pyramid(30), 4)      
        self.assertEqual(pyramid(55), 5)       
    
    def test_large_valid_pyramids(self):
        self.assertEqual(pyramid(91), 6)       
        self.assertEqual(pyramid(140), 7)     
        self.assertEqual(pyramid(204), 8)     
        self.assertEqual(pyramid(285), 9)      
        self.assertEqual(pyramid(385), 10)     
    
    def test_impossible_numbers(self):
        self.assertEqual(pyramid(2), "It is impossible")
        self.assertEqual(pyramid(3), "It is impossible")
        self.assertEqual(pyramid(4), "It is impossible")
        self.assertEqual(pyramid(6), "It is impossible")
        self.assertEqual(pyramid(7), "It is impossible")
        self.assertEqual(pyramid(8), "It is impossible")
        self.assertEqual(pyramid(15), "It is impossible")
        self.assertEqual(pyramid(100), "It is impossible")
        self.assertEqual(pyramid(200), "It is impossible")
```

```python
def is_balanced(number):

    if number < 0:
        raise ValueError("–ß–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º")
    
    digits_str = str(number)
    n = len(digits_str)
    
    if n <= 2:
        return True
    
    if n % 2 == 1:  
        mid_index = n // 2
        left_sum = sum(int(d) for d in digits_str[:mid_index])
        right_sum = sum(int(d) for d in digits_str[mid_index + 1:])
    else:  
        mid_start = n // 2 - 1
        mid_end = n // 2 + 1
        left_sum = sum(int(d) for d in digits_str[:mid_start])
        right_sum = sum(int(d) for d in digits_str[mid_end:])
    
    return left_sum == right_sum

#–¢–µ—Å—Ç—ã
import unittest

class TestIsBalancedFunction(unittest.TestCase):
    
    def test_single_digit_numbers(self):
        self.assertTrue(is_balanced(0))
        self.assertTrue(is_balanced(1))
        self.assertTrue(is_balanced(5))
        self.assertTrue(is_balanced(9))
    
    def test_two_digit_numbers(self):
        self.assertTrue(is_balanced(10))
        self.assertTrue(is_balanced(11))
        self.assertTrue(is_balanced(99))
        self.assertTrue(is_balanced(50))
    
    def test_balanced_odd_length(self):
        self.assertTrue(is_balanced(121))    
        self.assertTrue(is_balanced(12321))    
        self.assertTrue(is_balanced(13631)) 
        self.assertTrue(is_balanced(101))     
        self.assertTrue(is_balanced(1234321)) 
    
    def test_balanced_even_length(self):
        self.assertTrue(is_balanced(1221))     
        self.assertTrue(is_balanced(123321))   
        self.assertTrue(is_balanced(1001))     
        self.assertTrue(is_balanced(2332))     
        self.assertTrue(is_balanced(4554))     
```
